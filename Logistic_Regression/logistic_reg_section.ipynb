```python
#df[column list].apply(your_fun , axis=1)
def clean_age(colum):
    age = colum[0]
    p_class = colum[1]
    
    if pd.isnull(age):
        if p_class == 1 :
            return 38
        elif p_class == 2:
            return 29
        else:
            return 25
    else:
        return age
    
titanic_df['Age'] = titanic_df[['Age', 'Pclass']].apply(clean_age, axis=1)
```

```python
from sklearn.model_selection import train_test_split

X = titanic_df.iloc[: , 1:]
y = titanic_df['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression, Perceptron, SGDClassifier
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
svm_classifier = make_pipeline(StandardScaler(), SGDClassifier())
perceptron_classifier = Perceptron()
lm_classifier = LogisticRegression(C=10)
```

```python
svm_classifier.fit(X_train, y_train)
perceptron_classifier.fit(X_train, y_train)
lm_classifier.fit(X_train, y_train)
```

```python
y_predict_svm_classifier = svm_classifier.predict(X_test)
y_predict_perceptron = perceptron_classifier.predict(X_test)
y_predict = lm_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
sns.heatmap(confusion_matrix(y_test, y_predict), annot=True)
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
```

```python

print("Logistic Regression")
print(classification_report(y_test, y_predict))
```

```python
def roc_plot(roc_p, lw):
    plt.figure()
    lw = lw
    for elm in roc_p:
        plt.plot(
            elm["fpr"],
            elm["tpr"],
            lw=lw,
            label="ROC curve (area = %0.2f)" % elm["roc_auc"],
        )
    plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("Receiver operating characteristic example")
    plt.legend(loc="lower right")
    plt.show()
```

```python
y_score = lm_classifier.decision_function(X_test)
y_score_svm = svm_classifier.decision_function(X_test)
y_score_perceptron = perceptron_classifier.decision_function(X_test)
classifiers = [y_score_svm]
roc_data = []
for y_score_ in classifiers:
    temp = {}
    temp["fpr"], temp["tpr"], _ =  roc_curve(y_test, y_score_)
    temp["roc_auc"] = roc_auc = auc(temp["fpr"], temp["tpr"])
    roc_data.append(temp)
roc_plot(roc_data, 4)
```
