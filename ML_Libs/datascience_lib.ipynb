```python
def random_walk(walk, step):
    for i in range(step):
        temp_walk = 1 if random.randint(0,1) else -1
        walk.append(temp_walk)
    
    res_list = [0]
    for index in range(1, len(walk)):
        res_list.append(walk[index] + res_list[index-1])
    
    
    return res_list
```

```python
position = 0
walk = [position]

random_walk(walk, 5000)

print(walk)
total_walk = random_walk(walk, 5000)
total_walk
```

## Random

```python
def mean_khodam(arr_):
    return np.sum(arr_)/arr_.size

def tozi_normal_khodemun(*dimension, mean=0, variance=1):
    data = np.random.randn(*dimension) 
    return (variance ** 0.5) * data + mean

mean_khodam(tozi_normal_khodemun(4, 2, 8, 3, mean=10, variance=15))
```

### Conditional Logic as Array Operations

The numpy.where function is a vectorized version of the ternary expression x if condition else y

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
```

```python
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
```

```python
cond = np.array([True, False, True, True, False])
```

```python
result = np.where(cond, xarr, yarr)
result
```

```python
random_walk = np.random.randn(5 * 10 ** 5)
```

```python
random_walk = np.where(random_walk > 0, 1, -1)
random_walk
```

```python
random_walk[(random_walk != 1 ) & (random_walk != -1)]
```
