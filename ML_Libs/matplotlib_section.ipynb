```python
# Reference : Data Science from Scratch Book


variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
```

```python
plt.plot(xs, variance, 'g-', label='variance') # green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2') # red dot-dashed line
plt.plot(xs, total_error, 'b:', label='total error') # blue dotted line
# Because we've assigned labels to each series,
# we can get a legend for free (loc=9 means "top center")
plt.legend(loc=0)
plt.xlabel("model complexity")
plt.ylabel("val")

plt.xticks(range(8))
# plt.yticks(bias_squareds_squared)
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

```python
iris_df = pd.read_csv('data/iris.csv')
```

```python
iris_df.head()
```

```python
new_sepal_less = iris_df[iris_df['sepal_length'] < 4.5 ]['sepal_length']
new_sepal_bet = iris_df[(4.5 <= iris_df['sepal_length']) & (iris_df['sepal_length'] < 5) ]['sepal_length']
new_sepal_mor = iris_df[iris_df['sepal_length'] >= 5 ]['sepal_length']
```

```python
X = iris_df['sepal_length']
y = iris_df['sepal_width']

plt.figure(figsize=(9, 10))


plt.subplot(231)
plt.scatter(X, y)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')

plt.title('sepal l vs sepal w')




plt.subplot(232)
plt.hist(new_sepal_less, bins=50)
plt.xlabel('val')
plt.ylabel('count')

plt.title('sepal less than 4.5')



plt.subplot(233)
plt.hist(new_sepal_bet, bins=20)
plt.xlabel('val')
plt.ylabel('count')

plt.title('sepal between 4.5 and 5')


plt.subplot(234)
plt.hist(new_sepal_mor, bins=20)
plt.xlabel('val')
plt.ylabel('count')

plt.title('sepal more than 5')
```

```python
X = iris_df['sepal_length']
y = iris_df['sepal_width']

plt.figure(figsize=(9,40))


plt.subplot2grid((10,2),(3,1))
plt.scatter(X, y)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')

plt.title('sepal l vs sepal w')




plt.subplot2grid((10,2),(7,1))
plt.hist(new_sepal_less, bins=50)
plt.xlabel('val')
plt.ylabel('count')

plt.title('sepal less than 4.5')



plt.subplot2grid((10,2),(9,1))
plt.hist(new_sepal_bet, bins=20)
plt.xlabel('val')
plt.ylabel('count')

plt.title('sepal between 4.5 and 5')


plt.subplot2grid((10,2),(5,0))
plt.hist(new_sepal_mor, bins=20)
plt.xlabel('val')
plt.ylabel('count')

plt.title('sepal more than 5')
```
