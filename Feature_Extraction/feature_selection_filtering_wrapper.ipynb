```python
from sklearn.datasets import make_classification
from sklearn.feature_selection import f_classif, mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import pandas as pd
```

```python
data = make_classification(n_samples=1000, n_features=50, n_classes=3, n_informative=3)
```

```python
X, y = data
cross_val_score(RandomForestClassifier(n_estimators=50, max_depth=2), X, y).mean()
```

```python
data_df_f_im  = pd.DataFrame(X)
data_df_f_im["target"] = y
```

```python
rf_classifier = RandomForestClassifier(n_estimators=50, max_depth=2)
rf_classifier.fit(X,y)
f_im = rf_classifier.feature_importances_
for idx, val in enumerate(f_im):
    if val < 0.005:
        data_df_f_im.drop(idx, axis=1, inplace=True)
data_df_f_im
```

```python
cross_val_score(RandomForestClassifier(n_estimators=50, max_depth=2), data_df_f_im.iloc[:, :-1].values, data_df_f_im["target"].values).mean()
```

```python
data_df = pd.DataFrame(X)
data_df["target"] = y
```

```python
data_df_mu = pd.DataFrame(X)
data_df_mu["target"] = y
```

```python
f_val, p_val = f_classif(X, y)
for idx, val in enumerate(f_val):
    if val < 1:
        data_df.drop(idx, axis=1, inplace=True)
data_df
```

```python
X_selected = data_df.iloc[:, :-1].values
y_selected = data_df["target"].values
```

```python
cross_val_score(RandomForestClassifier(n_estimators=50, max_depth=2), X_selected, y_selected).mean()
```

```python
mi = mutual_info_classif(X, y)
for idx, val in enumerate(mi):
    if val < 0.01:
        data_df_mu.drop(idx, axis=1, inplace=True)
data_df_mu
```

```python
X_selected = data_df_mu.iloc[:, :-1].values
y_selected = data_df_mu["target"].values
```

```python
cross_val_score(RandomForestClassifier(n_estimators=50, max_depth=2), X_selected, y_selected).mean()
```

```python
from sklearn.feature_selection import SelectKBest, RFE
```

```python
select_cls = SelectKBest(mutual_info_classif, k=22)
select_cls.fit_transform(X,y)
```

```python
selector = RFE(RandomForestClassifier(n_estimators=50, max_depth=2), n_features_to_select=22)
X_selected = selector.fit_transform(X, y)
```

```python
cross_val_score(RandomForestClassifier(n_estimators=50, max_depth=2), X_selected, y_selected).mean()
```
